## Set up

```python
import pandas as pd
```

## Input Arguments

```python
DATASET_DP = f"../data/yelp_dataset/sample/sample_100_biz"
```

## Load data

```python
review_df = pd.read_parquet(f"{DATASET_DP}/yelp_academic_dataset_review_sample_100_biz.parquet")
```

```python
review_df
```

## Load business

```python
# load businesses
business_df = pd.read_parquet(f'{DATASET_DP}/yelp_academic_dataset_business_sample_100_biz.parquet')
```

```python
business_df
```

## Join Business & Review

```python
denom_df = (
    pd.merge(
        # TODO: fix duplicated columns
        review_df.rename(columns={'stars': 'review_stars'}),
        business_df.add_prefix("biz_").rename(columns={"biz_business_id": "business_id"}),
        how="left",
        on="business_id",
        validate="m:1"
    )
)
denom_df
```

```python
denom_df.describe(include="all").T
```

## Persist data

```python
DENOM_REVIEW_FD = f"{DATASET_DP}/denom_review.parquet"
denom_df.to_parquet(DENOM_REVIEW_FD, index=False)
```
